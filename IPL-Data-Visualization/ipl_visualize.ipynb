### Some basic Questions

#### 1. Which team won the most matches?

```python
match_wins=df['match_winner'].value_counts()
sns.barplot(y=match_wins.index,x=match_wins.values,palette='mako')
```

#### 2. Toss decision trends

```python
sns.countplot(x=df['toss_decision'],palette='mako')
```

#### 3. Toss decision VS Match Winner

```python
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(count*100)/df.shape[0]
percentage.round(2)
```

#### 4. How do teams win? (Runs vs Wickets)

```python
sns.countplot(x=df['won_by'],palette='mako')
```

#### 5. Most player of the match awards

```python
count=df['player_of_the_match'].value_counts().head(10)
sns.barplot(y=count.index,x=count.values,palette='mako')
```

#### 6. Top 2 scorers

```python
high=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
sns.barplot(y=high.index,x=high.values,palette='mako')
```

#### 7. 10 Best bowling figure

```python
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets']=df['highest_wickets'].astype(int)

bowling=df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)

sns.barplot(y=bowling.index,x=bowling.values,palette='mako')
```
